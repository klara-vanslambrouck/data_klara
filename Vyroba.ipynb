```python
import pandas as pd
```

Kolik režisérů na seriálu pracovalo?

```python
director = pd.read_csv("data/FRIENDS_DIRECTORS.csv")
f"Na seriálu pracovalo celkem {director["DIRECTOR"].nunique()} režisérů."
```

Kolik scénáristů na seriálu pracovalo?

```python
writer = pd.read_csv("data/FRIENDS_WRITERS.csv")
f"Na seriálu pracovalo celkem {writer["WRITER"].nunique()} scénáristů." 
```

Který režisér natočil nejvíce dílů?

```python
top_director = director["DIRECTOR"].value_counts().idxmax()
ep_count = director["DIRECTOR"].value_counts().max()
f"Nejvíce epizod natočil {top_director}. Celkem natočil {ep_count} epizod."
```

Který scénárista se podílel na nejvíce epizodách?

```python
top_writer = writer["WRITER"].value_counts().idxmax()
w_ep_count = writer["WRITER"].value_counts().max()
f"Nejvíce epizod napsal {top_writer}.Celkem se podílel na {w_ep_count} epizodách."
```

Epizody kterého režiséra měly lepší hodnocení od diváků?

```python
info = pd.read_csv("data/FRIENDS_INFO_CLEAN.csv")
merge = pd.merge(info, director,how="inner",on="EPISODE_ID")
avg_rating = merge.groupby("DIRECTOR")["imdb_rating"].mean()
rating_sort = avg_rating.sort_values(ascending=False)

best_director = rating_sort.head(1).idxmax()
best_avg_rating = rating_sort.head(1).max()

f"Režisérem nejlépe hodnocených epizod je {best_director} s průměrným hodnocením {best_avg_rating}."
```

Epizody kterého režiséra měly nejhorší hodnocení od diváků?

```python
merge = pd.merge(info, director,how="inner",on="EPISODE_ID")
avg_rating = merge.groupby("DIRECTOR")["imdb_rating"].mean()
rating_sort_worst = avg_rating.sort_values(ascending=True)

worst_director= rating_sort_worst.head(1).idxmax()
worst_avg_rating = rating_sort_worst.head(1).max()

f"Režisérem nejhůře hodnocených epizod je {worst_director} s průměrným hodnocením {worst_avg_rating}."
```

Epizody kterého scénáristy měly lepší hodnocení diváků?

```python
w_merge = pd.merge(info, writer,how="inner",on="EPISODE_ID")
w_avg_rating = w_merge.groupby("WRITER")["imdb_rating"].mean()
w_rating_sort = w_avg_rating.sort_values(ascending=False)

best_writer = w_rating_sort.head(1).idxmax()
best_w_rating = w_rating_sort.head(1).max()

f"Scénáristou nejlépe hodnocených epizod je {best_writer} s průměrným hodnocením {best_w_rating}."
```

Epizody kterého scénáristy měly nejhorší hodnocení od diváků?

```python
w_merge = pd.merge(info, writer,how="inner",on="EPISODE_ID")
w_avg_rating = w_merge.groupby("WRITER")["imdb_rating"].mean()
w_rating_sort_worst = w_avg_rating.sort_values(ascending=True)

worst_writer= w_rating_sort_worst.head(1).idxmax()
worst_w_avg_rating = w_rating_sort_worst.head(1).max()

f"Scénáristou nejhůře hodnocených epizod je {worst_writer} s průměrným hodnocením {worst_w_avg_rating}."
```
