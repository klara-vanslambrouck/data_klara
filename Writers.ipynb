```python
import pandas as pd
```

```python
data = pd.read_csv("Data/friends_writers_temp.csv")
```

```python
data['WRITER3']=data['WRITER3'].str.replace("\n", " ", regex=False)
```

```python
all_writers_list=list(data['WRITER3'].apply(
    lambda x: x.replace(r'\n','').split(' & ')
).values)
```

```python
all_writers_list
```

```python
unique_writers = list(set([
    w
    for w_list in all_writers_list
    for w in w_list
]))
```

```python
# split between lowercase and uppercase letters
import re
split_pattern = re.compile(
    r'(?<=[a-z])(?=[A-Z])(?<!Mc)(?<!Mac)(?<!-)'
)

# parsed_names = []
# for full in unique_writers:
#     parts = split_pattern.split(full)
#     parsed_names.extend([p.strip() for p in parts if p.strip()])


# for name in parsed_names:
#     print(name)

# clean each sublist
cleaned_all_writers_list = [
    [p.strip() for name in sublist for p in split_pattern.split(name) if p.strip()]
    for sublist in all_writers_list
]

# optional: remove duplicates inside each sublist
cleaned_all_writers_list = [list(dict.fromkeys(lst)) for lst in cleaned_all_writers_list]


cleaned_all_writers_list
```

```python
data['WRITER']=cleaned_all_writers_list
```

```python
data_new=data.explode('WRITER').drop(columns=['WRITER3'])
data_new
```

```python
data_new.to_csv("Data/Friends_writers.csv",index=False)
```
