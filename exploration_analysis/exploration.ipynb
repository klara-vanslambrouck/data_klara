```python
script_messy[script_messy["speaker"].str.contains("scene directions", case=False, na=False)].shape[0]
script_messy[script_messy["speaker"].str.contains("scene directions", case=False, na=False)]
```

```python
script_clean = script_messy[~script_messy["speaker"].str.strip().str.lower().eq("scene directions")]
script_clean.to_csv("script_clean.csv", index=False) # uloží novou tabulku bez řádků se "Scene Directions"
```

```python
script_clean = pd.read_csv("script_clean.csv")
```

```python
# uložím si hlavní postavy do seznamu
main_characters = ["Monica Geller", "Chandler Bing", "Rachel Green", "Ross Geller", "Joey Tribbiani", "Phoebe Buffay", "#ALL#"]

script_clean = script_clean.dropna(subset=["speaker"]) # odstraním řádky s NaN ve sloupci "speaker"

# vytvořím nový DataFrame bez hlavních postav
script_no_main = script_clean[~script_clean["speaker"].str.contains("|".join(main_characters), case=False, na=False)]

# display(script_no_main)

# odstraním duplicitní řádky, aby byl výpis přehlednější
script_no_main.drop_duplicates(subset=["speaker"])

number_of_side_characters = script_no_main["speaker"].nunique()
print(f"The total number of side characters: {number_of_side_characters}.")
```

```python
# zobrazím 20 nejčastějších vedlejších postav
script_no_main["speaker"].value_counts().head(20)
```

```python
# spočítá počet replik pro každou postavu
counts = script_no_main["speaker"].value_counts()

# nechá si jen ty, které mají víc než 1 repliku
valid_speakers = counts[counts > 1].index

script_no_main = script_no_main[script_no_main["speaker"].isin(valid_speakers)]

number_of_valid_spekares = script_no_main["speaker"].nunique()

print(f"The total number of side characters with more than 1 line: {number_of_valid_spekares}.")
```

```python
# Počet řádků v čistém skriptu -> tzn. počet replik (řádků) bez "Scene Directions"
script_clean.shape[0]
```

```python
main_characters_2 = ["Monica Geller", "Chandler Bing", "Rachel Green", "Ross Geller", "Joey Tribbiani", "Phoebe Buffay"]

# spočítá počet replik pro každou hlavní postavu
script_clean[script_clean["speaker"].isin(main_characters)]["speaker"].value_counts()
```

```python
# spočítá celkový počet replik hlavních postav
main_total = script_clean["speaker"].isin(main_characters_2).sum()
print("Main characters total lines:", main_total)
```

```python
# spočítá kolik replik bylo v jaké sezóně, seřadí od nejméně po nejvíce
script_clean["season"].value_counts(ascending=True)
```

```python
# počet replik v jednotlivých epizodách
episode_counts = script_clean["EPISODE_ID"].value_counts()

# najde epizodu s nejvíce a nejméně replikami
max_episode = episode_counts.idxmax()
min_episode = episode_counts.idxmin()

print(f"Episode with most lines: {max_episode} – {episode_counts.max()} lines")
print(f"Episode with fewest lines: {min_episode} – {episode_counts.min()} lines")
```

```python
# počet replik v jednotlivých epizodách
episode_lines = script_clean.groupby("EPISODE_ID").size().reset_index(name="num_lines")

episode_data = pd.merge(episode_lines, friends_info[["EPISODE_ID", "imdb_rating"]], on="EPISODE_ID")

correlation = episode_data["num_lines"].corr(episode_data["imdb_rating"])
print("Korelace mezi počtem replik a hodnocením:", round(correlation, 3))

```

```python
import matplotlib.pyplot as plt

plt.scatter(episode_data["num_lines"], episode_data["imdb_rating"])
plt.xlabel("Počet replik v epizodě")
plt.ylabel("IMDb hodnocení")
plt.title("Vztah mezi počtem replik a hodnocením epizody")
plt.show()
```
